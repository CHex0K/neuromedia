```python
# -*- coding: utf-8 -*-
from pathlib import Path

from google.colab import drive

PROJECT_DIR = Path("/content/neuromedia")
DRIVE_DIR = Path("/content/drive/MyDrive/neuromedia")
INPUT_DIR = DRIVE_DIR / "input"
SUPPORTED_EXTENSIONS = {
    ".txt",
    ".wav", ".mp3", ".flac", ".ogg",
    ".mp4", ".avi", ".mkv", ".mov", ".webm",
}

drive.mount("/content/drive")
DRIVE_DIR.mkdir(parents=True, exist_ok=True)
INPUT_DIR.mkdir(parents=True, exist_ok=True)

INPUT_FILES = sorted(
    path for path in INPUT_DIR.rglob("*")
    if path.is_file() and path.suffix.lower() in SUPPORTED_EXTENSIONS
)

if not INPUT_FILES:
    raise FileNotFoundError(
        f"No supported input files found in {INPUT_DIR}. "
        "Upload or sync files there, for example: "
        "MyDrive/neuromedia/input/4590080h.mp4"
    )

print("Input files:")
for input_file in INPUT_FILES:
    print(f"- {input_file}")

OUTPUT_ROOT = DRIVE_DIR / "outputs" / "colab_t4"
TRIBE_CACHE_DIR = DRIVE_DIR / "cache" / "tribev2"
NIMARE_DATA_DIR = DRIVE_DIR / "cache" / "nimare"
DECODER_CACHE = NIMARE_DATA_DIR / "correlation_decoder_top50_fthr005.pkl.gz"
SURFACE_CACHE = NIMARE_DATA_DIR / "surface_term_maps_fsaverage5_top50_fthr005.npz"

for path in (OUTPUT_ROOT, TRIBE_CACHE_DIR, NIMARE_DATA_DIR):
    path.mkdir(parents=True, exist_ok=True)

print(f"OUTPUT_ROOT={OUTPUT_ROOT}")
print(f"TRIBE_CACHE_DIR={TRIBE_CACHE_DIR}")
print(f"NIMARE_DATA_DIR={NIMARE_DATA_DIR}")
```

```python
# -*- coding: utf-8 -*-
import json
import os
import shlex
import subprocess
import sys
import time
from pathlib import Path

def make_output_dir(input_file):
    relative = input_file.relative_to(INPUT_DIR)
    safe_parts = list(relative.with_suffix("").parts)
    return OUTPUT_ROOT.joinpath(*safe_parts)

LOG_DIR = DRIVE_DIR / "logs"
STATE_DIR = DRIVE_DIR / "run_state"
LOG_DIR.mkdir(parents=True, exist_ok=True)
STATE_DIR.mkdir(parents=True, exist_ok=True)

timestamp = time.strftime("%Y%m%d_%H%M%S")
log_path = LOG_DIR / f"tribe_nimare_{timestamp}.log"
pid_path = STATE_DIR / "tribe_nimare.pid"
runner_path = PROJECT_DIR / "run_batch_tribe_nimare.py"

commands = []
for input_file in INPUT_FILES:
    output_dir = make_output_dir(input_file)
    output_dir.mkdir(parents=True, exist_ok=True)
    cmd = [
        sys.executable,
        str(PROJECT_DIR / "tribe_nimare_interpreter.py"),
        str(input_file),
        "--output-dir", str(output_dir),
        "--cache-dir", str(TRIBE_CACHE_DIR),
        "--nimare-data-dir", str(NIMARE_DATA_DIR),
        "--decoder-cache", str(DECODER_CACHE),
        "--surface-cache", str(SURFACE_CACHE),
        "--device", "cuda",
        "--n-cores", "1",
        "--frequency-threshold", "0.05",
        "--max-features", "50",
        "--min-feature-study-fraction", "0.01",
        "--max-feature-study-fraction", "0.9",
        "--top-k", "50",
    ]
    if (output_dir / "tribe_activity_fsaverage5.npy").is_file():
        cmd.append("--reuse-tribe")
    commands.append(cmd)

runner_payload = {
    "commands": commands,
}
runner_payload_json = json.dumps(runner_payload, ensure_ascii=False)
runner_code = """# -*- coding: utf-8 -*-
import json
import subprocess
import sys
import time

runner_payload = json.loads(r'''__PAYLOAD_JSON__''')
commands = runner_payload["commands"]

for index, cmd in enumerate(commands, start=1):
    print(f\"\\n[{index}/{len(commands)}] Running: {cmd[2]}\", flush=True)
    print(\"Command: \" + \" \".join(cmd), flush=True)
    started = time.time()
    result = subprocess.run(cmd)
    elapsed = time.time() - started
    print(f\"Finished with exit code {result.returncode} in {elapsed / 60:.1f} min\", flush=True)
    if result.returncode != 0:
        sys.exit(result.returncode)

print(\"All files processed.\", flush=True)
""".replace("__PAYLOAD_JSON__", runner_payload_json)
runner_path.write_text(runner_code, encoding="utf-8")

env = os.environ.copy()
env["PYTHONUTF8"] = "1"
env["OMP_NUM_THREADS"] = "1"
env["MKL_NUM_THREADS"] = "1"
env["OPENBLAS_NUM_THREADS"] = "1"
env["NUMEXPR_NUM_THREADS"] = "1"

with log_path.open("ab") as log_file:
    process = subprocess.Popen(
        [sys.executable, str(runner_path)],
        stdout=log_file,
        stderr=subprocess.STDOUT,
        env=env,
        start_new_session=True,
    )

pid_path.write_text(str(process.pid), encoding="utf-8")
(STATE_DIR / "tribe_nimare.log_path").write_text(str(log_path), encoding="utf-8")

print(f"Started background TRIBE/NiMARE job. PID: {process.pid}")
print(f"Log: {log_path}")
print(f"PID file: {pid_path}")
print("Run the next monitor cell to check progress.")
```

```python
# -*- coding: utf-8 -*-
import os
import subprocess
from pathlib import Path

STATE_DIR = Path("/content/drive/MyDrive/neuromedia/run_state")
pid_path = STATE_DIR / "tribe_nimare.pid"
log_path_file = STATE_DIR / "tribe_nimare.log_path"

if not pid_path.exists() or not log_path_file.exists():
    raise FileNotFoundError("No background run state found. Run the previous cell first.")

pid = int(pid_path.read_text(encoding="utf-8").strip())
log_path = Path(log_path_file.read_text(encoding="utf-8").strip())

alive = subprocess.run(
    ["bash", "-lc", f"ps -p {pid} -o pid=,etime=,cmd="],
    capture_output=True,
    text=True,
    check=False,
)

print(f"PID: {pid}")
print(f"Log: {log_path}")
print("Status:")
print(alive.stdout.strip() or "not running")
print("\nLast 80 log lines:")
if log_path.exists():
    tail = subprocess.run(
        ["tail", "-n", "80", str(log_path)],
        capture_output=True,
        text=True,
        check=False,
    )
    print(tail.stdout)
else:
    print("Log file does not exist yet.")
```
